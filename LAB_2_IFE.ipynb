```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Covariance matrix and expected returns (already defined)

# Define the range of target returns
target_returns = np.linspace(0.000120, 0.000140, num=50)

# Initialize lists to store portfolio returns and risks
portfolio_returns = []
portfolio_risks = []

# Step 2: Mean-variance optimization for each target return
for target_return in target_returns:
    # Objective function: minimize portfolio risk
    def portfolio_risk(weights, cov_matrix):
        return np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

    # Constraint: sum of weights equals 1 and portfolio return matches target return
    def constraint_sum_of_weights_and_return(weights, expected_returns, target_return):
        return np.sum(weights) - 1, np.dot(expected_returns, weights) - target_return

    # Bounds for weights (between 0 and 1)
    bounds = [(0, 1)] * len(expected_returns)

    # Define initial guess (equal weights)
    initial_guess = np.ones(len(expected_returns)) / len(expected_returns)

    # Perform mean-variance optimization
    result = minimize(portfolio_risk, initial_guess, args=(cov_matrix,),
                      constraints={'type': 'eq', 'fun': constraint_sum_of_weights_and_return, 'args': (expected_returns, target_return)},
                      bounds=bounds)

    # Extract optimal weights
    weights = result.x

    # Calculate portfolio return and risk
    portfolio_return = np.dot(expected_returns, weights)
    portfolio_returns.append(portfolio_return)
    portfolio_risk = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    portfolio_risks.append(portfolio_risk)

# Step 3: Plot the efficient frontier
plt.figure(figsize=(10, 5))
plt.plot(portfolio_risks, portfolio_returns, label='Efficient Frontier', color='blue')
plt.xlabel('Portfolio Risk (Standard Deviation)')
plt.ylabel('Portfolio Return')
plt.title('Markowitz Efficient Frontier')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt


# Define the efficient frontier function
def efficient_frontier(weights, cov_matrix, expected_returns):
    portfolio_return = np.dot(expected_returns, weights)
    portfolio_risk = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    return portfolio_risk, portfolio_return

# Define risk tolerance levels for the two points
target_risk_1 = 0.000100
target_risk_2 = 0.00200

# Initialize lists to store portfolio returns and risks
portfolio_returns = []
portfolio_risks = []

# Step 2: Mean-variance optimization for each target risk level
for target_risk in [target_risk_1, target_risk_2]:
    # Objective function: maximize portfolio return for a given level of risk
    def portfolio_return_objective(weights, cov_matrix, expected_returns):
        portfolio_risk, portfolio_return = efficient_frontier(weights, cov_matrix, expected_returns)
        return -portfolio_return  # Minimize negative return (maximize return)

    # Constraint: sum of weights equals 1 and portfolio risk equals target risk
    def constraint_sum_of_weights_and_risk(weights, cov_matrix, expected_returns, target_risk):
        portfolio_risk, _ = efficient_frontier(weights, cov_matrix, expected_returns)
        return np.sum(weights) - 1, portfolio_risk - target_risk

    # Bounds for weights (between 0 and 1)
    bounds = [(0, 1)] * len(expected_returns)

    # Define initial guess (equal weights)
    initial_guess = np.ones(len(expected_returns)) / len(expected_returns)

    # Perform mean-variance optimization
    result = minimize(portfolio_return_objective, initial_guess, args=(cov_matrix, expected_returns,),
                      constraints={'type': 'eq', 'fun': constraint_sum_of_weights_and_risk, 'args': (cov_matrix, expected_returns, target_risk)},
                      bounds=bounds)

    # Extract optimal weights
    weights = result.x

    # Calculate portfolio return and risk
    portfolio_return = np.dot(expected_returns, weights)
    portfolio_returns.append(portfolio_return)
    portfolio_risk = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    portfolio_risks.append(portfolio_risk)

# Plot the efficient frontier
plt.figure(figsize=(10, 6))
plt.plot(portfolio_risks, portfolio_returns, 'ro', label='Optimal Portfolios')
plt.xlabel('Portfolio Risk (Standard Deviation)')
plt.ylabel('Portfolio Return')
plt.title('Efficient Frontier with Two Risk Tolerance Levels')
plt.legend()
plt.grid(True)
plt.show()


```
